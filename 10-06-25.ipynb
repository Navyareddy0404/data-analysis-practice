```python
# exercise
As a teacher, analyze exam scores of students to understand their performance:

1. Collect exam scores into a Python list. For example: [22, 45, 60, 75, 80].

2. Create a histogram of these scores to visualize the distribution.
```

```python
exam_scores = [22, 45, 60, 75, 80]
bins=[0, 20, 40, 60, 80, 100]
plt.hist(exam_scores, bins=bins ,edgecolor='k')

plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.title('Distribution of Exam Scores')
plt.show()
```

```python
study_hours=[2,3,1,4,3,5,2,6,5,7,20,10,50,60,70]
exam_scores=[65,75,60,80,70,85,70,90,88,92,100,50,60,100,90]

plt.scatter(study_hours,exam_scores,label='student data')
plt.xlabel('study hours')
plt.xlabel('exam scores')
plt.title('study hours vs exam scores')
plt.show()
```
